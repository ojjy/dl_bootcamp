```python
import random
import evaluate
import numpy as np

from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

## Dataset ì¤€ë¹„

ê·¸ ë‹¤ìŒ ê°ì • ë¶„ì„ì„ ìœ„í•´ ì‚¬ìš©í•  imdb datasetì„ `load_dataset` í•¨ìˆ˜ë¡œ ë‹¤ìš´ë¡œë“œ ë°›ìŠµë‹ˆë‹¤.

```python
imdb = load_dataset("imdb")
imdb
```

`load_dataset`ì€ HuggingFaceì˜ `datasets` libraryì˜ í•¨ìˆ˜ë¡œ, HuggingFaceì˜ hubì—ì„œ datasetì„ ë‹¤ìš´ë¡œë“œ ë°›ì„ ìˆ˜ ìˆë„ë¡ ë§Œë“  í•¨ìˆ˜ì…ë‹ˆë‹¤.
ì¶œë ¥ ê²°ê³¼ë¥¼ ë³´ì‹œë©´ `imdb`ëŠ” `train`, `test`, ê·¸ë¦¬ê³  `unsupervised` dataë¡œ êµ¬ì„±ë˜ì–´ìˆìŠµë‹ˆë‹¤.
ì´ ì¤‘ì—ì„œ ìš°ë¦¬ëŠ” `train`ê³¼ `test`ë¥¼ í™œìš©í•©ë‹ˆë‹¤.

`train` dataë¥¼ í•œ ë²ˆ ì‚´í´ë³´ê² ìŠµë‹ˆë‹¤.

```python
imdb['train'][0]
```

`train`ê³¼ `test`ì˜ ê° dataëŠ” `text`ì™€ `label`ë¡œ êµ¬ì„±ë˜ì–´ìˆìŠµë‹ˆë‹¤.
ê°ê°ì€ ì˜í™”ë¦¬ë·°ì™€ í•´ë‹¹ ì˜í™”ë¦¬ë·°ì˜ ê¸ì •/ë¶€ì • ì—¬ë¶€ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.
ì´ëŠ” ì´ì „ ì£¼ì°¨ë“¤ì—ì„œ ì‚¬ìš©í•œ imdb datasetê³¼ ë™ì¼í•©ë‹ˆë‹¤.

ì´ë²ˆì—ëŠ” tokenizerë¥¼ ë¶ˆëŸ¬ì™€ì„œ ë¯¸ë¦¬ textë“¤ì„ tokenizeí•©ë‹ˆë‹¤.

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

def preprocess_function(data):
    return tokenizer(data["text"], truncation=True)

imdb_tokenized = imdb.map(preprocess_function, batched=True)
```

Tokenizerë¥¼ ì‹¤í–‰í•  ë•Œ ë„˜ê²¨ì£¼ì—ˆë˜ `truncation` ì˜µì…˜ì€ ì£¼ì–´ì§„ textê°€ ì¼ì • ê¸¸ì´ ì´ìƒì´ë©´ ì˜ë¼ë‚´ë¼ëŠ” ì˜ë¯¸ì…ë‹ˆë‹¤.
ë§Œì•½ íŠ¹ì • ê¸¸ì´ ê°’ì´ ê°™ì´ ì£¼ì–´ì§€ì§€ ì•ŠëŠ”ë‹¤ë©´ `bert-base-cased`ë¥¼ í•™ìŠµí•  ë•Œ ì‚¬ìš©í•œ textì˜ ìµœëŒ€ ê¸¸ì´ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ê°’ì„ ê²°ì •í•©ë‹ˆë‹¤.

ë§ˆì§€ë§‰ ì¶œë ¥ ê²°ê³¼ë¥¼ ë³´ë©´, `text`ì™€ `label` ì´ì™¸ì— `input_ids`ê°€ ìƒê¸°ì‹  ê²ƒì„ í™•ì¸í•˜ì‹¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
ì´ëŠ” ìš°ë¦¬ê°€ `AutoTokenizer.from_pretrained`ë¡œ ë¶ˆëŸ¬ì˜¨ tokenizerë¡œ textë¥¼ tokenë“¤ë¡œ ë‚˜ëˆ„ê³  ì •ìˆ˜ indexë¡œ ë³€í™˜í•œ ê²°ê³¼ì…ë‹ˆë‹¤.

ì´ë²ˆì—ëŠ” `train` dataë¥¼ ìª¼ê°œ training dataì™€ validation dataë¥¼ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤.

```python
imdb_split = imdb_tokenized['train'].train_test_split(test_size=0.2)
imdb_train, imdb_val = imdb_split['train'], imdb_split['test']
imdb_test = imdb_tokenized['test']
```

## Model êµ¬í˜„

ì´ë²ˆì—ëŠ” text ë¶„ë¥˜ë¥¼ ìˆ˜í–‰í•  Transformerë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.
ì´ì „ì—ëŠ” Transformerì˜ êµ¬ì„± ìš”ì†Œë“¤ì„ ì§ì ‘ êµ¬í˜„í•˜ì—¬ í•©ì³¤ìŠµë‹ˆë‹¤.
ì´ë²ˆì—ëŠ” HuggingFaceì˜ BERTë¥¼ í™œìš©í•˜ì—¬ ì¸ìë§Œ ë„˜ê²¨ì£¼ëŠ” ì‹ìœ¼ë¡œ êµ¬í˜„í•´ë³´ê² ìŠµë‹ˆë‹¤:

```python
from transformers import BertConfig

config = BertConfig()

config.hidden_size = 64  # BERT layerì˜ ê¸°ë³¸ hidden dimension
config.intermediate_size = 64  # FFN layerì˜ ì¤‘ê°„ hidden dimension
config.num_hidden_layers = 2  # BERT layerì˜ ê°œìˆ˜
config.num_attention_heads = 4  # Multi-head attentionì—ì„œ ì‚¬ìš©í•˜ëŠ” head ê°œìˆ˜
config.num_labels = 2  # ë§ˆì§€ë§‰ì— ì˜ˆì¸¡í•´ì•¼ í•˜ëŠ” ë¶„ë¥˜ ë¬¸ì œì˜ class ê°œìˆ˜

model = AutoModelForSequenceClassification.from_config(config)
```

BERTëŠ” ì´ì „ì— ë°°ìš´ Transformerì˜ architectureë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
ê·¸ë˜ì„œ BERTì˜ ì˜µì…˜ë“¤ë§Œ ìˆ˜ì •í•˜ë©´ vanilla Transformerë¥¼ ì‰½ê²Œ êµ¬í˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

Transformer êµ¬í˜„ ì´ì™¸ì— ë¶„ë¥˜ ë¬¸ì œì— ë§ì¶° ì²« ë²ˆì§¸ tokenì„ linear classifierë¥¼ ê±°ì¹˜ëŠ” ë“±ì˜ ê³¼ì •ì€ `AutoModelForSequenceClassification`ì´ êµ¬í˜„í•´ì¤ë‹ˆë‹¤.
ì¦‰, ìš°ë¦¬ê°€ `config`ë¡œ ë„˜ê²¨ì£¼ëŠ” BERTì˜ ë§ˆì§€ë§‰ì— linear classifierë¥¼ ë‹¬ì•„ì£¼ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.

## í•™ìŠµ ì½”ë“œ

ë‹¤ìŒì€ ìœ„ì—ì„œ êµ¬í˜„í•œ Transformerë¥¼ imdbë¡œ í•™ìŠµí•˜ëŠ” ì½”ë“œë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.
ë¨¼ì € ë‹¤ìŒê³¼ ê°™ì´ í•™ìŠµ ì¸ìë“¤ì„ ì •ì˜í•©ë‹ˆë‹¤.

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='hf_transformer',  # ëª¨ë¸, log ë“±ì„ ì €ì¥í•  directory
    num_train_epochs=10,  # epoch ìˆ˜
    per_device_train_batch_size=128,  # training dataì˜ batch size
    per_device_eval_batch_size=128,  # validation dataì˜ batch size
    logging_strategy="epoch",  # Epochê°€ ëë‚  ë•Œë§ˆë‹¤ training loss ë“±ì„ logí•˜ë¼ëŠ” ì˜ë¯¸
    do_train=True,  # í•™ìŠµì„ ì§„í–‰í•˜ê² ë‹¤ëŠ” ì˜ë¯¸
    do_eval=True,  # í•™ìŠµ ì¤‘ê°„ì— validation dataì— ëŒ€í•œ í‰ê°€ë¥¼ ìˆ˜í–‰í•˜ê² ë‹¤ëŠ” ì˜ë¯¸
    eval_strategy="epoch",  # ë§¤ epochê°€ ëë‚  ë•Œë§ˆë‹¤ validation dataì— ëŒ€í•œ í‰ê°€ë¥¼ ìˆ˜í–‰í•œë‹¤ëŠ” ì˜ë¯¸
    save_strategy="epoch",  # ë§¤ epochê°€ ëë‚  ë•Œë§ˆë‹¤ ëª¨ë¸ì„ ì €ì¥í•˜ê² ë‹¤ëŠ” ì˜ë¯¸
    learning_rate=1e-3,  # optimizerì— ì‚¬ìš©í•  learning rate
    load_best_model_at_end=True  # í•™ìŠµì´ ëë‚œ í›„, validation dataì— ëŒ€í•œ ì„±ëŠ¥ì´ ê°€ì¥ ì¢‹ì€ ëª¨ë¸ì„ ì±„íƒí•˜ê² ë‹¤ëŠ” ì˜ë¯¸
)
```

ê°ê°ì˜ ë¶€ë¶„ë“¤ì€ ì´ì „ ì£¼ì°¨ì—ì„œ ë°°ì› ë˜ ë‚´ìš©ë“¤ì„ ì„¤ì •í•˜ëŠ” ê²ƒì— ë¶ˆê³¼í•˜ë‹¤ëŠ” ê²ƒì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
ìš”ì•½í•˜ë©´ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:
- `epochs`: training dataë¥¼ ëª‡ ë²ˆ ë°˜ë³µí•  ê²ƒì¸ì§€ ê²°ì •í•©ë‹ˆë‹¤.
- `batch_size`: training dataë¥¼ ì–¼ë§ˆë‚˜ ì˜ê²Œ ì˜ë¼ì„œ í•™ìŠµí•  ê²ƒì¸ì§€ ê²°ì •í•©ë‹ˆë‹¤.
- `learning_rate`: optimizerì˜ learning rateë¥¼ ì–¼ë§ˆë¡œ í•  ê²ƒì¸ì§€ ê²°ì •í•©ë‹ˆë‹¤.
ìœ„ì˜ ë¶€ë¶„ë“¤ ì´ì™¸ì—ë„ ì‚¬ì†Œí•œ êµ¬í˜„ ìš”ì†Œë“¤ë„ ì§€ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë‹¤ìŒì€ loss ì´ì™¸ì˜ í‰ê°€ í•¨ìˆ˜ë“¤ì„ êµ¬í˜„í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤.

```python
import evaluate

accuracy = evaluate.load("accuracy")


def compute_metrics(pred):
    predictions, labels = pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

`evaluate` ë˜í•œ HuggingFaceì˜ libraryë¡œ ë‹¤ì–‘í•œ í‰ê°€ í•¨ìˆ˜ë“¤ì„ ì œê³µí•˜ê³  ìˆìŠµë‹ˆë‹¤.
ì´ë²ˆ ì‹¤ìŠµì˜ ê²½ìš°, ê°ì • ë¶„ì„ ë¬¸ì œëŠ” ë¶„ë¥˜ ë¬¸ì œì´ê¸° ë•Œë¬¸ì— ì •í™•ë„ë¥¼ ê³„ì‚°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
ìœ„ì™€ ê°™ì´ ì˜ˆì¸¡ ê²°ê³¼(`pred`)ì™€ ì‹¤ì œ label(`labels`)ê°€ ì£¼ì–´ì¡Œì„ ë•Œ ì •í™•ë„ë¥¼ ê³„ì‚°í•˜ëŠ” ê²ƒì€ `evaluate`ì˜ accuracy í•¨ìˆ˜ë¡œ êµ¬í˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë§ˆì§€ë§‰ìœ¼ë¡œ ìœ„ì˜ ìš”ì†Œë“¤ì„ ì¢…í•©í•˜ì—¬ í•™ìŠµí•  ìˆ˜ ìˆëŠ” `Trainer`ë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.

```python
from transformers import EarlyStoppingCallback


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=imdb_train,
    eval_dataset=imdb_val,
    compute_metrics=compute_metrics,
    tokenizer=tokenizer,
    # callbacks = [EarlyStoppingCallback(early_stopping_patience=1)]
)
```

ëª¨ë¸, training ì¸ì, trainingê³¼ validation data, ë¶€ê°€ì ì¸ í‰ê°€ í•¨ìˆ˜, ê·¸ë¦¬ê³  tokenizerë¥¼ ë„˜ê²¨ì£¼ë©´ ëì…ë‹ˆë‹¤.
ë³„ê°œë¡œ early stoppingê³¼ ê°™ì€ ê¸°ëŠ¥ë„ ì£¼ì„ ì¹œ ë¶€ë¶„ê³¼ ê°™ì´ `callbacks`ë¡œ êµ¬í˜„í•  ìˆ˜ ìˆìœ¼ë‹ˆ ì°¸ê³ í•´ì£¼ì‹œê¸¸ ë°”ëë‹ˆë‹¤.

ìœ„ì™€ ê°™ì´ ë§Œë“  `Trainer`ëŠ” ë‹¤ìŒê³¼ ê°™ì´ í•™ìŠµì„ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
trainer.train()
```

ë³´ì‹œë‹¤ì‹œí”¼ training lossëŠ” ì˜ ë–¨ì–´ì§€ëŠ” ë°˜ë©´, validation lossëŠ” ì¤‘ê°„ë¶€í„° ì­‰ ì˜¬ë¼ê°€ëŠ” ê²ƒì„ ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
Overfittingì´ ì¼ì–´ë‚¬ë‹¤ê³  ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ìœ„ì™€ ê°™ì´ í•™ìŠµì´ ëë‚œ í›„ validation lossê°€ ê°€ì¥ ë‚®ì€ ëª¨ë¸ì„ ê°€ì§€ê³  test dataì˜ ì„±ëŠ¥ì„ í‰ê°€í•˜ëŠ” ê²ƒì€ ë‹¤ìŒê³¼ ê°™ì´ êµ¬í˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
trainer.evaluate(imdb_test)
```

ì´ì „ì— í•™ìŠµ ì¸ìì—ì„œ `load_best_model_at_end=True`ë¥¼ ë„˜ê²¨ì¤¬ê¸° ë•Œë¬¸ì— `trainer`ëŠ” í•™ìŠµì´ ëë‚œ í›„, ê¸°ë³¸ì ìœ¼ë¡œ validation lossê°€ ê°€ì¥ ì¢‹ì€ ëª¨ë¸ì„ ê°€ì§€ê³  `evaluate`ë¥¼ ì§„í–‰í•©ë‹ˆë‹¤.
ì‹¤ì œë¡œ ê²°ê³¼ë¥¼ ë³´ë©´ `eval_loss`ê°€ ê°€ì¥ ë‚®ì€ validation lossì™€ ìœ ì‚¬í•œ ê²ƒì„ ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

í‰ê°€í•  ë•Œ ì‚¬ìš©í•œ ëª¨ë¸ì€ ë‹¤ìŒê³¼ ê°™ì´ ì €ì¥í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
trainer.save_model()
```

ì—¬ê¸°ì„œëŠ” ì˜í™” ë¦¬ë·°ê°€ ì£¼ì–´ì¡Œì„ ë•Œ, labelì´ 0(ë¶€ì •ì ì¸ ë¦¬ë·°)ì¸ì§€ 1(ê¸ì •ì ì¸ ë¦¬ë·°)ì¸ì§€ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë³´ì—¬ì¤„ ë¿ë§Œ ì•„ë‹ˆë¼ ê·¸ ì‹ ë¢°ë„ë¥¼ `score`ë¡œ ë„˜ê²¨ì£¼ê²Œ ë©ë‹ˆë‹¤.

ì´ì²˜ëŸ¼ HuggingFaceë¥¼ í™œìš©í•˜ë©´ ëª¨ë¸ì´ë‚˜ ì˜ˆì¸¡, í•™ìŠµ ì½”ë“œë¥¼ êµ¬í˜„í•  í•„ìš” ì—†ì´ ì¸ìë¡œ ì„¤ì •ê°’ë“¤ë§Œ ë„˜ê²¨ì£¼ë©´ ì‰½ê²Œ êµ¬í˜„ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

## Transfer learning

ì´ë²ˆì—ëŠ” taskëŠ” ì˜í™” ë¦¬ë·° ê°ì • ë¶„ì„ìœ¼ë¡œ ìœ ì§€í•˜ë˜, ëª¨ë¸ì„ distilbertë¥¼ fine-tuningí•˜ëŠ” ê²ƒìœ¼ë¡œ ë°”ê¿”ë³´ê² ìŠµë‹ˆë‹¤.
ëª¨ë¸ì€ ë‹¤ìŒê³¼ ê°™ì´ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
id2label = {0: "NEGATIVE", 1: "POSITIVE"}
label2id = {"NEGATIVE": 0, "POSITIVE": 1}

model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert/distilbert-base-uncased", num_labels=2, id2label=id2label, label2id=label2id
)
```

`id2label`ê³¼ `label2id`ëŠ” ì˜ˆì¸¡ ê²°ê³¼ì˜ ê°€ë…ì„±ì„ ìœ„í•´ì„œ ì¶”ê°€í•˜ì˜€ìŠµë‹ˆë‹¤.
ê¸°ì¡´ê³¼ ê°™ì´ `AutoModelForSequenceClassification`ë¥¼ ì‚¬ìš©í•˜ê³  ìˆìŠµë‹ˆë‹¤.
ë‹¬ë¼ì§„ ì ì€ `from_config`ê°€ ì•„ë‹Œ `from_pretrained`ë¥¼ ì‚¬ìš©í•œë‹¤ëŠ” ê²ƒì…ë‹ˆë‹¤.
`from_pretrained`ë¥¼ ì‚¬ìš©í•˜ë©´ HuggingFace hubì— ìˆëŠ” pre-trained ëª¨ë¸ë“¤ì„ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
ê°ì • ë¶„ì„ ë¬¸ì œì˜ class ìˆ˜ì— ë§ì¶° `num_labels`ë¥¼ 2ë¡œ ì„¤ì •í•˜ë©´ ëª¨ë¸ êµ¬í˜„ì€ ê±°ì˜ ëë‚¬ìŠµë‹ˆë‹¤.

ì´ë²ˆì—ëŠ” ë§ˆì§€ë§‰ layerë¥¼ ì œì™¸í•œ parameterë“¤ì„ freezeí•´ë³´ê² ìŠµë‹ˆë‹¤.
ê·¸ ì „ì— freeze í•´ì•¼ í•˜ëŠ” layerë“¤ì„ í™•ì¸í•˜ê¸° ìœ„í•´ `model`ì„ ì¶œë ¥í•´ë³´ê² ìŠµë‹ˆë‹¤.

```python
model
```

ë³´ì‹œë‹¤ì‹œí”¼ `distilbert`ëŠ” ê¸°ì¡´ì˜ distilbert ëª¨ë¸ì— í•´ë‹¹í•˜ê³  ë‚˜ë¨¸ì§€ `pre_classifier`, `classifier`ëŠ” text ë¶„ë¥˜ë¥¼ ìœ„í•´ ìƒˆë¡­ê²Œ ì¶”ê°€ëœ layerë“¤ì…ë‹ˆë‹¤.
ì¦‰, ë‹¤ìŒê³¼ ê°™ì´ `distilbert`ì— í•´ë‹¹í•˜ëŠ” parameterë“¤ë§Œ freezeí•˜ë©´ ë©ë‹ˆë‹¤.

```python
for param in model.distilbert.parameters():
  param.requires_grad = False
```

ì´ëŸ¬ë©´ ëª¨ë¸ êµ¬í˜„ì€ ì™„ì „íˆ ë§ˆì³¤ìŠµë‹ˆë‹¤.
ë‹¤ìŒì€ distilbertë¥¼ pre-trainí•  ë•Œ ì‚¬ìš©í–ˆë˜ tokenizerë¥¼ ë¶ˆëŸ¬ì˜¤ê³ , ì´ tokenizerë¥¼ ê°€ì§€ê³  ì´ì „ê³¼ ë˜‘ê°™ì´ imdb datasetë¥¼ ì „ì²˜ë¦¬í•©ë‹ˆë‹¤.

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilbert-base-uncased")

def preprocess_function(data):
    return tokenizer(data["text"], truncation=True)

imdb_tokenized = imdb.map(preprocess_function, batched=True)
imdb_split = imdb_tokenized['train'].train_test_split(test_size=0.2)
imdb_train, imdb_val = imdb_split['train'], imdb_split['test']
imdb_test = imdb_tokenized['test']
```

ë‚˜ë¨¸ì§€ëŠ” ê¸°ì¡´ì˜ í•™ìŠµê³¼ì •ê³¼ ì™„ì „íˆ ë™ì¼í•©ë‹ˆë‹¤. `training_args`ëŠ” ê¸°ì¡´ê³¼ ë˜‘ê°™ì´ í™œìš©í•  ë•Œ distilbertë¥¼ fine-tuningí•˜ëŠ” ì½”ë“œëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=imdb_train,
    eval_dataset=imdb_val,
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)

trainer.train()
trainer.save_model()
```

í•™ìŠµí•œ ëª¨ë¸ì— ëŒ€í•œ ê²°ê³¼ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.

```python
trainer.predict(imdb_test)
```
